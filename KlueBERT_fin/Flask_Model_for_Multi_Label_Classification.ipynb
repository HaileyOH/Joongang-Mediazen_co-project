```python
# 기존 모듈spq 
from flask import Flask, render_template, request, jsonify, send_file, make_response
from flask_ngrok import run_with_ngrok
from urllib.parse import unquote

from transformers import AutoTokenizer, BertForSequenceClassification
import torch
import numpy as np
from io import BytesIO


# 모듈 import-추가
import time
import sys 
import os 
import re

import pandas as pd
from selenium import webdriver   # 실행 전 selenium 설치 필요
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.action_chains import ActionChains
import seaborn as sns
from collections import Counter
import plotly 
import plotly.graph_objs as go 

import urllib.request as req
from bs4 import BeautifulSoup
import base64

import warnings
warnings.filterwarnings('ignore')



#----

model_path = "/content/drive/MyDrive/projects/KlueBERT/model/checkpoint-best-model"
tokenizer = AutoTokenizer.from_pretrained(model_path)
model = BertForSequenceClassification.from_pretrained(model_path)

app = Flask(__name__, template_folder=TEMPLATE, static_folder=STATIC)
run_with_ngrok(app)


#----


# chrome_options = webdriver.ChromeOptions()
# chrome_options.add_argument('--headless')
# chrome_options.add_argument('--no-sandbox')
# chrome_options.add_argument('--disable-dev-shm-usage')
# chrome_options.add_argument("user-agent=Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/102.0.5005.63 Safari/537.36")
# driver = webdriver.Chrome('chromedriver', chrome_options=chrome_options)


@app.route("/")
def home():
  return render_template("index.html")

@app.route("/_input_name")
def input_name():
    cafe = unquote(request.args.get('key_data', "", type=str)).strip()
    review_list = get_review(cafe)  # 크롤링하기
    prediction, chart = predict_from_input2(review_list)
    return jsonify(crawl_result=render_template("live_results_review.html", 
                                                review_predict=prediction, review=review_list, chart=chart))

@app.route("/_input_helper")  
def input_helper():
  text = unquote(request.args.get("text_data", "", type=str)).strip()
  prediction = predict_from_input(text)
  return jsonify(result=render_template('live_results.html',predict=prediction))


def get_review(cafe):   ## 바로 검색(시연용)
    cafe_csv = pd.read_csv('/content/drive/MyDrive/projects/KlueBERT/cafe_review_DB_revised_merged.csv', encoding="utf-8")
    review_list = []
    for i in range(len(cafe_csv)):
      if cafe_csv['cafe'][i] == cafe:
          review_list.append(cafe_csv['review'][i])
    return review_list

# def get_review(cafe):   ## 크롤링
#         #pdb.set_trace()
#     cafe_csv = pd.read_csv('/content/drive/MyDrive/projects/KlueBERT/url_completed.csv', encoding='cp949')
#     cafe_url = cafe_csv.loc[cafe_csv['title'] == cafe, 'url'].values[0]
#         #pdb.set_trace()
#     try:
#         driver.get(cafe_url) 
#         # for i in range(10):
#         while True:
#             try:
#                 for i in range(3):
#                     more_reviews = driver.find_element_by_css_selector('a._3iTUo')
#                     more_reviews.click()
#                     time.sleep(0.1)
#             except:
#                 break
#         reviews = driver.find_elements_by_css_selector('div.faZHB > a > span.WoYOw')
#         review_list = list(set([a.text for a in reviews]))
#         # print("-"*10,"Tag 크롤링 종료", "-"*10)
#         return review_list
#     except IndexError: 
#         print('none')


def predict_from_input(text):
  threshold = 0.1
  token = tokenizer(text, return_tensors="pt")
  outputs = model(**token)
  logits = outputs.logits
  sigmoid = torch.nn.Sigmoid()
  probs = sigmoid(logits.squeeze().cpu())
  predictions = np.zeros(probs.shape)
  predictions[np.where(probs >= threshold)] = 1
  predicted_labels = [model.config.id2label[idx] for idx, label in enumerate(predictions) if label == 1.0]
  plot(predicted_labels)
  return predicted_labels


# 기존 리뷰 재라벨 출력
def predict_from_input2(text):
  threshold = 0.1
  label_dictionary = {}
  label_list = []
  for batch in text:
    threshold = 0.1
    token = tokenizer(batch, return_tensors="pt", padding = True, truncation = True, max_length = 128)
    outputs = model(**token)
    logits = outputs.logits
    sigmoid = torch.nn.Sigmoid()
    probs = sigmoid(logits.squeeze().cpu())
    predictions = np.zeros(probs.shape)
    predictions[np.where(probs >= threshold)] = 1
    predicted_labels = [model.config.id2label[idx] for idx, label in enumerate(predictions) if label == 1.0]
    label_dictionary[batch] = [predicted_labels]
    label_list.extend(predicted_labels)
  chart = plot(label_list)
  return label_dictionary, chart


def plot(predicted_labels):
  fig,axis= plt.subplots(1)
  #concatlist = []         # 데이터 준비 :   다중 리스트 -> 하나의 리스트
  #for i in predicted_labels:
  #    concatlist +=i
  count = Counter(predicted_labels)
  top_tags = []
  top_count = []
  n = 8
  for i in range(n):     # 빈도수 가장많은 5개 (태그 개수가 5이하면 전부 다 출력)
      if i <len(count.most_common(n=n)):
          top_tags.append(count.most_common(n=n)[i][0])
          top_count.append(count.most_common(n=n)[i][1])
      else:
          break
    # 그리기
  fig = plt.figure(figsize=(8,8))
  fig.set_facecolor('white')
  ax = fig.add_subplot()
  pie = ax.pie(top_count,labels = top_tags, ## 파이차트 출력
        startangle=90, ## 시작점을 90도(degree)로 지정
        counterclock=False, ## 시계 방향으로 그린다.
        autopct=lambda p : '{:.2f}%'.format(p), ## 퍼센티지 출력
        wedgeprops=dict(width=0.5) ## 중간의 반지름 0.5만큼 구멍을 뚫어준다.
        )
  plt.legend(pie[0],top_tags) ## 범례 표시

  # 그려진 img 파일 내용을 html 랜더링 쪽에 전송
  img = BytesIO()
  fig.savefig(img)
  img.seek(0)
  plt.savefig('/content/drive/MyDrive/projects/KlueBERT/img/new_plot1.png') # 그래프 new_plot1.png로 static에 저장
  chart = base64.b64encode(img.getvalue()).decode('ascii')
  time.sleep(5) #저장할 시간 기다림
  return chart



#app.debug = True
app.run()
```
