## 영문, 숫자, 특수문자, 이모지 삭제

### 함수정의

```python
def rfWords(inputdf, col_name):
    inputdf[col_name] = inputdf[col_name].replace('[^가-힣]+', ' ', regex=True)   # 영문, 숫자, 특수문자 삭제
    emoji_pattern = re.compile("["            # 이모지 삭제
                               u"\U0001F600-\U0001F64F"  # emoticons
                               u"\U0001F300-\U0001F5FF"  # symbols & pictographs
                               u"\U0001F680-\U0001F6FF"  # transport & map symbols
                               u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
                               "]+", flags=re.UNICODE)
    df_li = inputdf[col_name].tolist()
    rf_li = []
    for i in df_li:
        # emoji 삭제
        tokens = re.sub(emoji_pattern, ' ', i)
        # multi-space 삭제
        tokens_re = re.sub('\s{2,}', ' ' ,tokens).strip()
        # str_result += tokens
        rf_li.append(tokens_re)    
    refined_df = pd.DataFrame(rf_li)
    inputdf[col_name] = refined_df[0]
    return inputdf
```

### 삭제 및 확인

```python
rfWords(df, 'review')
df
```
